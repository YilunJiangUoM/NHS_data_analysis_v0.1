# Month/Year: 01/2020

Exclusion: 

            1. Home Visit & Unknown
            2. More than 28 Days & Unknown / Data Issue  

```python
csv_name = 'CCG_CSV_Jan_20.csv'
file_path = folder_path + csv_name
dataframe = pd.read_csv(file_path)
dataframe.drop('CCG_CODE', axis=1, inplace=True)
dataframe.drop('CCG_ONS_CODE', axis=1, inplace=True)
dataframe.drop('STP_ONS_CODE', axis=1, inplace=True)
dataframe.drop('HCP_TYPE', axis=1, inplace=True)
#dataframe.drop('REGION_ONS_CODE', axis=1, inplace=True)
```

## Data filtering & Analysis

```python
# dna: did not attend
# f2f: face-to-face
# remote: Video Conference/Online & Telephone

remote_attended_tmp = dataframe[(dataframe['APPT_STATUS'] == 'Attended') 
                                & (dataframe['APPT_MODE'] != 'Home Visit') 
                                & (dataframe['APPT_MODE'] != 'Face-to-Face') 
                                & (dataframe['APPT_MODE'] != 'Unknown') 
                                & (dataframe['TIME_BETWEEN_BOOK_AND_APPT'] !='More than 28 Days')
                                & (dataframe['TIME_BETWEEN_BOOK_AND_APPT'] !='Unknown / Data Issue')
                                ].groupby('CCG_NAME').sum().reset_index()

remote_dna_tmp = dataframe[(dataframe['APPT_STATUS'] == 'DNA') 
                                & (dataframe['APPT_MODE'] != 'Home Visit') 
                                & (dataframe['APPT_MODE'] != 'Face-to-Face') 
                                & (dataframe['APPT_MODE'] != 'Unknown') 
                                & (dataframe['TIME_BETWEEN_BOOK_AND_APPT'] !='More than 28 Days')
                                & (dataframe['TIME_BETWEEN_BOOK_AND_APPT'] !='Unknown / Data Issue')
                                ].groupby('CCG_NAME').sum().reset_index()

f2f_attended_tmp = dataframe[(dataframe['APPT_STATUS'] == 'Attended') 
                            & (dataframe['APPT_MODE'] == 'Face-to-Face')
                            & (dataframe['TIME_BETWEEN_BOOK_AND_APPT'] !='More than 28 Days')
                            & (dataframe['TIME_BETWEEN_BOOK_AND_APPT'] !='Unknown / Data Issue')                    
                            ].groupby('CCG_NAME').sum().reset_index()

f2f_dna_tmp = dataframe[(dataframe['APPT_STATUS'] == 'DNA') 
                        & (dataframe['APPT_MODE'] == 'Face-to-Face') 
                        & (dataframe['TIME_BETWEEN_BOOK_AND_APPT'] !='More than 28 Days')
                        & (dataframe['TIME_BETWEEN_BOOK_AND_APPT'] !='Unknown / Data Issue')
                        ].groupby('CCG_NAME').sum().reset_index()


# drop na

remote_attended_tmp = remote_attended_tmp.dropna()
remote_dna_tmp = remote_dna_tmp.dropna()
f2f_attended_tmp = f2f_attended_tmp.dropna()
f2f_dna_tmp = f2f_attended_tmp.dropna()

#count, mean, max, min, 5th, 10th, 25th, 50th, 75th, 90th, 95th
print('-------------------------------------------------')
print('Data description of attended remote consultation in 2020/01:')
print(remote_attended_tmp
      .describe(percentiles=[.05, .10, .25, .50, .75, .90, .95]))
print('-------------------------------------------------')
print('Minimum:')
print(remote_attended_tmp.min())
print('-------------------------------------------------')
print('Maximum:')
print(remote_attended_tmp.max())
print('-------------------------------------------------')
print('CCG number:' + str(len(remote_attended_tmp)))
```
